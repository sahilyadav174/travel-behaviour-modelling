```python
df1=pd.get_dummies(df["What difficulties you may face if motorized vehicle free day is implemented?"],drop_first=True)

df.head()

df1.head()

df.head()

df['Age (in years)'].value_counts()

df2=pd.get_dummies(df['Age (in years)'],drop_first=True)

df2.head()

df=df.join(df2)

df.drop("Age (in years)",axis=1,inplace=True)

df['Preferred frequency of Motorized Vehicle Free Day'].value_counts()

df3=pd.get_dummies(df['Preferred frequency of Motorized Vehicle Free Day'],drop_first=True)

df=df.join(df3)

df.drop("Preferred frequency of Motorized Vehicle Free Day",axis=1,inplace=True)

df4=pd.get_dummies(df['Gender'],drop_first=True)

df=df.join(df4)

df.drop("Gender",axis=1,inplace=True)

df.head()

df5=pd.get_dummies(df['Occu'])

df5.drop(["Married Students","Project staff"],axis=1,inplace=True)

df=df.join(df5)

df.drop("Occu",axis=1,inplace=True)
```
